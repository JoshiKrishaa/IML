```python
#Write a NumPy program to implement following operation
#to add, subtract, multiply, divide arguments element-wise
#Krishaa Joshi
#216270307047
import numpy as np
print("Addition=",np.add(1,2))
print("Subraction=",np.subtract(2,2))
print("Multiplication=",np.multiply(2,3))
print("Division=",np.divide(4,2))

```

```python
#Write a NumPy program to implement following operation
#to round elements of the array to the nearest integer
#Krishaa Joshi
#216270307047
import numpy as np
a=np.array([2.7,3.6,4.5])
print("Array=",np.round(a))
```

```python
#Write a NumPy program to implement following operation
#to calculate mean across dimension, in a 2D numpy array
#Krishaa Joshi
#216270307047
import numpy as np
a=np.array([[1,2,3],[4,5,6]])
print("Array columnwise=",np.mean(a,axis=0))
print("Array rowwise=",np.mean(a,axis=1))


```
